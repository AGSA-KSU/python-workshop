# Thermal Units Exercise

```python
# Import modules
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_notebook
output_notebook()
```

```python
# Load tabular data
URL = 'https://raw.githubusercontent.com/soilwater/python-workshop/main/Day_1/ashland_bottoms_2019_2020.csv'
df = pd.read_csv(URL, comment='#')
df.head()
```

```python
# Print columsn in df
df.columns
```

```python
# CHeck type of timestamp
type(df['TIMESTAMP'][0])
```

```python
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'], format='%m/%d/%Y')
df.head()
```

```python
# Check type of timestamp
type(df['TIMESTAMP'][0])
```

```python
# Check for missing values represented as NaN (Not a Number)
df.isna() # False = 0 and True = 1
```

```python
# Count missing values
df.isna().sum()
```

```python
# Interpolate missing values
df['TEMP2MMIN'].interpolate(method='pchip', limit_area='inside', inplace=True)
df['TEMP2MMAX'].interpolate(method='pchip', limit_area='inside', inplace=True)
```

```python
# Check our work
df.isna().sum()
```

```python
f1 = figure(x_axis_type='datetime', plot_width=800, plot_height=400, title='Ashland Bottoms Weather')
f1.line(df['TIMESTAMP'], df['TEMP2MMIN'], name='Tmin', color='navy')
f1.line(df['TIMESTAMP'], df['TEMP2MMAX'], name='Tmax', color='tomato')
f1.yaxis.axis_label = 'Air Temperature (Celsius)'
show(f1)
```

```python
# Define growing season
planting_date = pd.to_datetime('15-Oct-2019', format='%d-%b-%Y')
harvest_date = pd.to_datetime('1-Jun-2020', format='%d-%b-%Y')
growing_season_length = harvest_date - planting_date
growing_season_length = growing_season_length.days + 1

print('Planting date', planting_date)
```

```python
# Select rows for growing season
idx_season = (df['TIMESTAMP'] >= planting_date) & (df['TIMESTAMP'] <= harvest_date)
idx_season[280:295]
```

```python
# Create a new Dataframe for the growing season
df_season = df.loc[idx_season, :]
df_season.head()
```

```python
# Reset index
df_season.reset_index(inplace=True, drop=True)
df_season.head()
```

```python
f2 = figure(x_axis_type='datetime', plot_width=800, plot_height=400, title='Ashland Bottoms Weather Growing Season')
f2.line(df_season['TIMESTAMP'], df_season['TEMP2MMIN'], name='Tmin', color='navy') 
# f2.line(source=df_season, x='TIMESTAMP', y='TEMP2MMIN', name='Tmin', color='navy')
f2.line(df_season['TIMESTAMP'], df_season['TEMP2MMAX'], name='Tmax', color='tomato')
f2.yaxis.axis_label = 'Air Temperature (Celsius)'
show(f2)
```

```python
# Find Tmax and Tmin for the growing season
Tmax = df_season['TEMP2MMAX'].max()
idx_Tmax = df_season['TEMP2MMAX'].idxmax()
day_Tmax = df_season.loc[idx_Tmax, 'TIMESTAMP']
print('The maximum temperature was', Tmax, 'Celsius', 'and occurred on',day_Tmax)

# Homework: Find Tmin and day of Tmin
```

```python
# Compute thermal time
T_base = 0 # Celsius
T_upper = 35 # Celsius
```

$$ if T_{avg} < T_{base}, \ then \ TU = T_{base} $$
$$ if T_{avg} > T_{upper}, \ then \ TU = T_{upper} $$
$$ TU = T_{avg} - T_{base}$$

```python
# Compute average temperature
T_avg = (df_season['TEMP2MMIN'] + df_season['TEMP2MMAX'])/2 # Element-wise computation

df_season.insert(4, 'TEMP2MAVG', T_avg)
df_season.head()
```

```python
TU_daily = (np.minimum(np.maximum(df_season['TEMP2MAVG'], T_base), T_upper))
TU_cumulative = TU_daily.cumsum()
df_season.insert(len(df_season.columns), 'TU_cumulative', TU_cumulative)
df_season.head()
```

```python
# Plot thermal units
f3 = figure(x_axis_type='datetime', plot_width=800, plot_height=400, title='Ashland Bottoms Weather Thermal Units')
f3.line(source=df_season, x='TIMESTAMP', y='TU_cumulative')
f3.yaxis.axis_label = 'Thermal Units C-day'
show(f3)
```

```python
# Find the total thermal units for the season
TU_total = df_season['TU_cumulative'].iloc[-1]
print('The total number of thermal units in the growing season was',round(TU_total))
```
